**Liệt kê các sản phẩm của nước Nhật theo mẫu sau, sắp tăng theo city**

```python
%%sql results <<
select suppliers.city,suppliers.companyName,products.productName,products.unitprice
from products inner join suppliers on products.supplierid=suppliers.id
where suppliers.country="Japan"
order by suppliers.city
```

```python
results
```

**Thống kê theo Số sản phẩm và Đơn giá trung bình được cung cấp theo từng quốc gia theo mẫu sau, dữ liệu được sắp tăng dần theo Số sản phẩm và giảm dần theo Đơn giá trung bình**

```python
%%sql results <<
SELECT t1.*, t2.count_city
from
(select suppliers.country as country,avg(products.unitprice) as avg_unitprice, count(products.id) as count_product
    from products inner join suppliers on products.supplierid=suppliers.id
    group by suppliers.country) as t1
join
(select country,count(city) as count_city
    from suppliers
    group by suppliers.country) as t2
on t1.country=t2.country
order by t1.count_product asc, t1.avg_unitprice desc
```

```python
results
```

**Liệt kê các đơn đặt hàng đặt trong tháng 3/2014 theo mẫu sau, dữ liệu sắp giảm dần theo orderdate và totalAmount**

```python
%%sql results <<
select customers.FirstName, customers.LastName, orders.OrderNumber, orders.OrderDate, orders.TotalAmount
from customers inner join orders on orders.CustomerId=customers.id
where year(orderdate)=2014 and month(orderdate)=3
order by orderdate desc, totalamount desc
```

```python
results
```

**Thống kê sản phẩm có tổng thành tiền nhỏ nhất trong mỗi năm theo mẫu sau**

```python
%%sql results <<
(select year(OrderDate) as "order_year", min(TotalAmount),
    ProductName
from products inner join orderitems on products.id =orderitems.productid inner join orders on orders.id = orderitems.orderid
where year(OrderDate) = 2012)
union
(select year(OrderDate) as "order_year", min(TotalAmount),
    ProductName
from products inner join orderitems on products.id =orderitems.productid inner join orders on orders.id = orderitems.orderid
where year(OrderDate) = 2013)
union
(select year(OrderDate) as "order_year", min(TotalAmount),
    ProductName
from products inner join orderitems on products.id =orderitems.productid inner join orders on orders.id = orderitems.orderid
where year(OrderDate) = 2014)
```

```python
results
```

**Liệt kê các đơn đặt hàng có ngày đặt hàng là ngày chủ nhật của tháng 3/2014 theo mẫu sau, dữ liệu sắp giảm theo orderdate**

```python
%%sql results <<
select orders.OrderNumber, 
    orders.OrderDate, 
    concat(customers.FirstName, " ",customers.LastName) as customer_name, 
    orders.TotalAmount
from customers inner join orders on orders.CustomerId=customers.id
where (year(orderdate)=2014) and (month(orderdate)=3) and DayofWeek(orders.OrderDate)='1'
order by orderdate desc
```

```python
results
```

**Thống kê số lần đặt hàng của các customer theo mẫu sau, chỉ in các customer nào có số lần đặt hàng >15, dữ liệu sắp giảm dần theo số lần đặt**

```python
%%sql results <<
select 
    concat(customers.FirstName, " ",customers.LastName) as customer_name,
    customers.Country,
    customers.City,
    count(orders.Id) as count_orders
from customers inner join orders on orders.CustomerId=customers.id
group by customers.Id
Having count(orders.Id) > 15
order by count(orders.Id) desc
```

```python
results
```

**2 quý nào có tổng thành tiền bán cao nhất**

```python
%%sql results <<
select 
    year(orderDate) as year ,QUARTER(orderDate) as quater, sum(TotalAmount) as totalamount
from orders
group by year(orderDate), QUARTER(orderDate)
order by sum(TotalAmount) desc
limit 2
```

```python
results
```

**Liệt kê thông tin của 10 đơn đặt hàng đặt mua sản phẩm có tên là Chai theo mẫu sau, dữ liệu sắp giảm theo TotalAmount**

```python
%%sql results <<
select orders.Id, orders.OrderDate, orders.OrderNumber, orders.CustomerId, orders.TotalAmount
from products inner join orderitems on products.Id=orderitems.ProductId inner join orders on orderitems.OrderId = orders.Id
where products.ProductName="Chai"
order by orders.TotalAmount desc
limit 10
```

```python
results
```

**Thay thế giá trị NULL (None) của cột sum_quantity thành giá trị 0**

```python
%%sql results <<
select ProductName,
        (case when Sum(i.Quantity) is null then 0 else Sum(i.Quantity) end) as sum_quantity
from products p left join orderitems i on p.Id=i.ProductId left join orders o on o.Id=i.OrderId
where p.isDiscontinued=0
Group by  p.ProductName
Order by sum_quantity;
```
